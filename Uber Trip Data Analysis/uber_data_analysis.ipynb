```
%pylab inline
import pandas
import seaborn
```

```
data = pandas.read_csv('/content/uber-raw-data-apr14.csv')
```

```
data.tail()
```

```
data['Date/Time']= data['Date/Time'].map(pandas.to_datetime)
```

```
data.tail()
```

```
def get_dom(dt):
  return dt.day
data['dom'] = data['Date/Time'].map(get_dom)
```

```
data.tail()
```

```
def get_weekday(dt):
  return dt.weekday()
data['weekday'] = data['Date/Time'].map(get_weekday)

def get_hour(dt):
 return dt.hour
data['hour'] = data['Date/Time'].map(get_hour)
```

```
data.tail()
```

```
hist(data.dom, bins=30, rwidth=10, range=(0.5, 30.5))
xlabel('date of the month')
ylabel('frequency')
title('frequency by DOM_UBER_APR2014')
```

```
def count_rows(rows):
  return len(rows)
by_date = data.groupby('dom').apply(count_rows)
by_date
```

```
bar(range(1,31),by_date)
```

```
by_date_sorted = by_date.sort_values()
by_date_sorted
```

```
bar(range(1,31),by_date_sorted)
xticks(range(1,31),by_date_sorted.index)
xlabel('date of the month')
ylabel('frequency')
title('frequency by DOM_UBER_APR2014')
```
